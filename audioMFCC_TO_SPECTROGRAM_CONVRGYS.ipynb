```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
import os
import matplotlib

import glob
import pylab
import librosa
import librosa.display
import numpy as np
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
```

```python
def wav_to_mfcc(filename):
    y, sr = librosa.load(filename,duration=4.0)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    fig = plt.Figure()
    canvas = FigureCanvas(fig)
    ax = fig.add_subplot(111)
    p = librosa.display.specshow(librosa.amplitude_to_db(mfccs, ref=np.max), ax=ax, y_axis=None, x_axis= None)
    fig.savefig(filename +'.png')
```

```python
files = glob.glob(r'C:\Users\dell15\concentrix\speaker3\*.wav')
#print (files)
#files[1]
for ele in files:
    wav_to_mfcc(ele)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
